## What each metric means

| Metric | Definition |
|---|---|
| `pct_addressed_by_male` | Share of lines *directed at* the protagonist that come from male characters (raw — cast-confounded) |
| `out_in_ratio` | Protagonist's outgoing dialogue ÷ incoming. >1 = initiator; <1 = responder |
| `bf_gender` | Gender of the protagonist's most-addressed alter (the character they address most by utterance count) |
| `bf_ratio_pct` | Share of protagonist's dialogue going to the most-addressed alter |
| `bf_reciprocity` | Whether the most-addressed-alter relationship is mutual |
| `isolation_pct` | Share of protagonist's lines addressed to self or non-characters (monologue) |

**Framing:** all social metrics are descriptive — not hypothesis tests.

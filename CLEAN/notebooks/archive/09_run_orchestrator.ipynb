# Notebook 09_run_orchestrator — Run N=17 Analysis Orchestrator

**Run this before:** `09_h1_homophily`, `09_extended_network_analysis`, `09b_keystone_dirw_experiment`

**What it does:** Runs `n17_orchestrator.py` which computes all analysis tables for N=17:
- H1 test battery (Pillar A)
- Addressee vs co-occurrence comparison including keystone table `step2f_keystone_agreement.csv` (Pillar B)
- Structural context (betweenness, quadrant typology, keystone)
- Descriptive extensions (female_alter_betw_z, burt_constraint, etc.)

**Output:** `analysis/h1_homophily/tables_n17/` + `figures_n17/`

**N=17 conventions:** Soul dropped, Sulley kept, one protagonist per film.

```python
from pathlib import Path
import subprocess, sys

CLEAN = Path('..').resolve()
script = CLEAN / 'analysis' / 'h1_homophily' / 'n17_orchestrator.py'

print('Running n17_orchestrator.py...')
result = subprocess.run(
    [sys.executable, str(script)],
    capture_output=True, text=True,
    cwd=str(CLEAN / 'analysis' / 'h1_homophily')
)
print(result.stdout[-3000:])
if result.returncode != 0:
    print('STDERR:', result.stderr[-1000:])
else:
    print('Done.')
```
